## Data Understanding

Dataset ini berisi semua transaksi yang terjadi di sebuah perusahaan ritel online yang berbasis di Inggris dan terdaftar secara resmi, yang tidak memiliki toko fisik (hanya berjualan online). Transaksi tersebut terjadi antara tanggal 1 Desember 2009 hingga 9 Desember 2011.

Berikut adalah rincian kolom-kolomnya:
* Invoice: Nomor faktur unik untuk setiap transaksi. Jika nomor ini diawali dengan 'C', itu menunjukkan pembatalan pesanan (cancellation).
* StockCode: Kode unik untuk setiap produk (item).
* Description: Nama atau deskripsi produk.
* Quantity: Jumlah produk yang dibeli dalam satu transaksi. Nilai negatif menunjukkan pengembalian barang.
* InvoiceDate: Tanggal dan waktu ketika transaksi (faktur) dibuat.
* Price: Harga satuan produk dalam poundsterling
* Customer ID: ID unik untuk setiap pelanggan. Nilai kosong (NaN) menunjukkan bahwa pelanggan tidak terdaftar atau transaksi dilakukan oleh tamu.
* Country: Negara tempat pelanggan berada.



Data Overview:  

- Data memiliki Baris: 1067371 dan Kolom:8
- Data memiliki 3 jenis kolom: string(5), integer(1), float(2)
- `Invoice` yang dimulai dengan 'C' perlu dibersihkan karena transaksinya dibatalkan
- Cek nilai negatif pada kolom `Quantity` dan `Price`

**Goals for RFM Analysis
Customer Segmentation (RFM Analysis)**
  
  khusus untuk Member (Hanya Transaksi yang terdapat CustomerID)

1. **Customer Segmentation (RFM Analysis)**:
   * Gunakan metrik Recency, Frequency, dan Monetary untuk mengelompokkan pelanggan ke dalam beberapa segmen (misalnya: Champions, At Risk, Potential Loyalists).
   * Identifikasi pelanggan bernilai tinggi serta pelanggan yang membutuhkan strategi reaktivasi.

2. **Customer Behavior Analysis**:
   * Analisis pola pembelian pelanggan berdasarkan frekuensi transaksi dan total pengeluaran.
   * Tentukan pelanggan yang paling sering berbelanja dan pelanggan yang sudah lama tidak bertransaksi untuk strategi retensi.

3. **Product Performance Insights**:
   * Analisis hubungan antara segmen pelanggan dan produk yang mereka beli untuk mengoptimalkan strategi penjualan.
   * Evaluasi produk yang sering dibeli oleh pelanggan bernilai tinggi agar dapat dijadikan fokus dalam promosi.

4. **Optimization Recommendations**:
   * Rekomendasikan kampanye pemasaran yang ditargetkan untuk masing-masing segmen pelanggan (misalnya: program loyalitas untuk Champions, diskon untuk At Risk).
   * Strategi upselling dan cross-selling berdasarkan preferensi pelanggan dari analisis RFM.

# Data Cleaning

# `Cek Missing Value`

```python
from IPython.display import display

# Cek Missing Value
print("Missing Value di Setiap Kolom:")
display(df.isnull().sum().to_frame(name="Missing Values"))

# Menghitung keseluruhan data tanpa missing value
def missing(dataset, key):
    tanpa_null = pd.DataFrame(len(dataset[key]) - dataset.isnull().sum(), columns=['Data Count'])
    percentage_null = pd.DataFrame((len(dataset[key]) - (len(dataset[key]) - dataset.isnull().sum())) / len(dataset[key]) * 100, columns=['Percentage'])
    percentage_null = percentage_null.round(2)

    # Menampilkan jumlah data tanpa missing value
    print("\nJumlah Data Tanpa Missing Value:")
    display(tanpa_null)

    # Menampilkan persentase nilai yang hilang
    print("\nPersentase Missing Value:")
    display(percentage_null)

    return tanpa_null, percentage_null

# Memanggil fungsi dan menampilkan hasilnya
tanpa_null, percentage_null = missing(df, 'Invoice')
```

# `Handling Missing Value`





Missing value akan di drop, karena bertujuan untuk RFM yang berfokus pada transaksi yang memiliki `CustomerID`

```python
#  Removing NaN's in Customer ID
print("Shape data sebelum menghapus NaN di CustomerID",df.shape)
df.dropna(subset="Customer ID",axis=0,inplace=True)
print("Shape data setelah menghapus NaN di CustomerID",df.shape)
```

```python
# Cek Missing Value setelah di handling
print("Missing values di setiap kolom setelah cleaning customerID :\n\n", df.isnull().sum())
```

### Menghapus transaksi yang dibatalkan dari Invoice

karena ada 2 tipe Invoice yaitu yang berawalan dengan "C" dan Tidak  
- berawalan C: Transaksi yang dibatalkan
- tanpa awalan C : Transaksi yang selesai

```python
# Melihat keseluruhan data Invoice
print(df[['Invoice']].to_string(index=False))
```

```python
# Melihat jumlah Invoice dengan awalan 'C' dan tanpa 'C'
cancelled_count = df[df['Invoice'].astype(str).str.startswith('C')]['Invoice'].count()
non_cancelled_count = df[~df['Invoice'].astype(str).str.startswith('C')]['Invoice'].count()

invoice_counts = pd.DataFrame({
"Invoice Type": ["Cancelled (C)", "Non-Cancelled"],
    "Count": [cancelled_count, non_cancelled_count]
})

print(invoice_counts.to_string(index=False))
```

Filter (Memilah) invoice tanpa awalan C

```python
# menggunakan operator (~) dapat memilih seluruh data di kolom Invoice column yang tidak mengandung "C".
df = df[~df.Invoice.str.contains('C',na=False)]
print("Data sudah bersih dari transaksi yang dibatalkan (berawalan C)")
```

```python
cancelled_count = df[df['Invoice'].astype(str).str.startswith('C')]['Invoice'].count()
non_cancelled_count = df[~df['Invoice'].astype(str).str.startswith('C')]['Invoice'].count()

invoice_counts = pd.DataFrame({
"Invoice Type": ["Cancelled (C)", "Non-Cancelled"],
    "Count": [cancelled_count, non_cancelled_count]
})

print(invoice_counts.to_string(index=False))
```
